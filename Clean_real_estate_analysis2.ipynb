### Merapikan file : kolom, format tanggal

```python
# rapikan nama kolom

df_clean.columns = (
    df_clean.columns
    .str.strip()
    .str.replace(" ", "_")
    .str.lower()
)

df_clean.columns
```

### Median Sale per year

```python
# median sale amount per year

yearly_trend = (
    df_clean
    .groupby("year")["sale_amount"]
    .median()
    .reset_index()
)

yearly_trend
```
